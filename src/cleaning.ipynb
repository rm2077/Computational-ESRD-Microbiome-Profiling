```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
```

```python
def normalize_data(data):
    df[data] = min_max_scaler.fit_transform(df[[data]])
```

```python

normalize_data("Urea")
```

```python
normalize_data("eGFR")
```

```python
study_group_mappings = {"Control" : 0,  "ESRD" : 1}
gender_mappings = {"Female": 0, "Male" : 1}
```

```python
df["Study.Group"] = df["Study.Group"].map(study_group_mappings)
```

```python
df["Gender"] = df["Gender"].map(gender_mappings)
```

```python
df
```

```python
cols_to_skip = ["Study.Group", "Age", "Gender", "Creatinine", "Urea", "eGFR"]
```

```python
cols_to_normalize = [col for col in df.columns if col not in cols_to_skip]
```

```python
df[cols_to_normalize] = min_max_scaler.fit_transform(df[cols_to_normalize])
```

```python
df.fillna(0)
```

```python
df.to_csv("data.csv", sep="\t")
```
